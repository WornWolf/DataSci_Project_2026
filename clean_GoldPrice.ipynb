## GoldPrice Managing

ใน GoldPrice ใช้ column - Date, Price(USD), Price(THB) 

1. แปลงวันที่ของ Date เป็น 1 ทั้งหมด

```python
goldPrice['Date'] = pd.to_datetime(goldPrice['Date'])
goldPrice['Date'] = goldPrice['Date'].dt.to_period('M').dt.to_timestamp()
```

```python
goldPrice.head()
```

2. สร้าง DataFrame เก็บข้อมูลเฉพาะที่ต้องการ
- ในที่นี้จะใช้ Date, Price(USD), Price(THB)

```python
newGoldPrice = goldPrice[['Date','United States(USD)','Thailand(THB)']]
newGoldPrice.rename(columns={'United States(USD)': 'Price(USD)', 'Thailand(THB)': 'Price(THB)'}, inplace=True)
```

```python
newGoldPrice.head()
```

```python
newGoldPrice.info()
```

3. New Column - Gold Change %
- New Price - Old Price / Old Price จะได้ % เป็น Price อันใหม่ที่เปลี่ยนไปกี่ %

```python
newGoldPrice["Gold Change %"] = newGoldPrice["Price(USD)"].pct_change() * 100
```

```python
newGoldPrice.head(3)
```

```python
sns.pairplot(newGoldPrice)
```

4. Export CSV

```python
newGoldPrice.to_csv('./dataset/cleaned_GoldPrice.csv', index=False)
```
